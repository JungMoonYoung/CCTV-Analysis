## 1. 분석 결과 종합

```python
# 주요 통계 요약
summary_stats = {
    '전체 자치구 수': len(df),
    '평균 방범용 CCTV': df['인구당_방범용'].mean().round(2),
    '평균 범죄율': df['인구당_CCTV효과범죄율'].mean().round(2),
    'CCTV-범죄율 상관계수': df['인구당_방범용'].corr(df['인구당_CCTV효과범죄율']).round(4),
    '우선순위 지역 수': len(df[df['분면'] == 'Q2: 저CCTV/고범죄 (우선순위)']),
    '효과적 지역 수': len(df[df['분면'] == 'Q4: 고CCTV/저범죄 (효과적)'])
}

print("="*70)
print("분석 결과 종합")
print("="*70)
for key, value in summary_stats.items():
    print(f"{key}: {value}")
```

## 2. 정책 제언 (우선순위별)

```python
recommendations = """
📋 정책 제언 (우선순위별)

【단기 과제 (6개월 이내)】

1. Q2 지역 긴급 CCTV 설치
   - 대상: 저CCTV/고범죄 자치구
   - 내용: 방범용 CCTV 집중 설치 (최소 중앙값 수준까지)
   - 우선 위치: 절도·강도 다발 지역, 어두운 골목길
   - 예상 효과: 범죄율 15-20% 감소 (회귀계수 기반 추정)

2. Q4 지역 모범 사례 분석
   - 대상: 고CCTV/저범죄 자치구
   - 내용: CCTV 배치 전략, 관제 시스템, 연계 방범 활동 조사
   - 활용: Q2 지역 설치 시 벤치마킹

【중기 과제 (1년 이내)】

3. Q1 지역 종합 방범 대책
   - 대상: 고CCTV/고범죄 자치구
   - 내용: CCTV 외 조명 개선, 경찰 순찰 강화, 주민 방범대 구성
   - 근거: CCTV만으로 범죄 억제 한계 (회귀 진단 결과)

4. 유형별 CCTV 재배치
   - 근거: 방범용 CCTV가 범죄율과 가장 높은 상관관계
   - 내용: 교통단속용·어린이안전용 일부를 방범용으로 전환 검토
   - 조건: 교통사고율·어린이 안전 지표 동시 고려 필요

【장기 과제 (2년 이상)】

5. 시계열 데이터 구축
   - 목적: CCTV 설치 전후 범죄율 변화 인과관계 검증
   - 방법: 3-5년 패널 데이터 수집, 이중차분법(DID) 분석
   - 활용: 정책 효과성 객관적 평가 근거 마련

6. 지능형 CCTV 도입
   - 내용: AI 기반 범죄 예측·실시간 알림 시스템
   - 우선 적용: Q2 지역 신규 설치 CCTV
   - 기대 효과: 범죄 예방 효과 극대화
"""

print(recommendations)
```

## 3. 연구의 한계점

```python
limitations = """
⚠️ 연구의 한계점

1. 인과관계 불명확
   - 문제: 횡단면 데이터(single-year)로 인과관계 검증 불가
   - 역인과성: 범죄 많은 곳에 CCTV를 더 설치했을 가능성
   - 해결 방안: 시계열 데이터 + 이중차분법(DID) 필요

2. 누락 변수 편향 (Omitted Variable Bias)
   - 미포함 요인: 조명 밝기, 경찰 순찰 빈도, 유동인구, 상권 밀집도
   - 영향: 회귀계수 과대/과소 추정 가능성
   - 해결 방안: 추가 통제변수 확보 필요

3. 공간적 자기상관 미고려
   - 문제: 인접 자치구 간 범죄율 영향 관계 미반영
   - 해결 방안: 공간회귀모형(Spatial Lag/Error Model) 적용

4. CCTV 품질 변수 부재
   - 문제: 해상도, 야간 촬영 성능, 관제 인력 등 질적 차이 미고려
   - 영향: CCTV 대수만으로 효과성 평가 한계

5. 범죄 암수 (Dark Figure of Crime)
   - 문제: 신고되지 않은 범죄 미포함
   - 영향: 실제 범죄율 과소 추정 가능성
"""

print(limitations)
```

## 4. 후속 연구 제안

```python
future_research = """
🔬 후속 연구 제안

1. 시계열 분석 (Time-Series Analysis)
   - 데이터: 2018-2024 CCTV 설치 및 범죄 발생 추이
   - 방법: 이중차분법(DID), 패널 고정효과 모형
   - 목표: CCTV 설치 인과효과 검증

2. 공간 분석 고도화
   - 데이터: 읍면동 단위 GeoJSON + 범죄 좌표
   - 방법: Hotspot 분석, Spatial Lag Model
   - 목표: CCTV 최적 배치 위치 도출

3. 기계학습 기반 예측 모형
   - 변수: CCTV, 인구, 유동인구, 상권, 조명, 경찰서 거리 등
   - 방법: Random Forest, XGBoost
   - 목표: 범죄 위험도 예측 및 CCTV 우선순위 자동화

4. 비용-편익 분석 (Cost-Benefit Analysis)
   - 비용: CCTV 설치·유지관리·관제 인력
   - 편익: 범죄 감소로 인한 사회적 비용 절감
   - 목표: 최적 투자 규모 산출

5. 질적 연구 병행
   - 방법: 주민 설문조사, 경찰 인터뷰
   - 내용: 범죄 두려움(Fear of Crime), CCTV 체감 효과
   - 목표: 정량 분석 보완 및 정책 수용성 제고
"""

print(future_research)
```
