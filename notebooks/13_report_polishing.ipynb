## 3. ë³´ê³ ì„œ ë‚´ ê·¸ë˜í”„ ë§í¬ ìœ íš¨ì„± ê²€ì‚¬

````python
import re

# ë³´ê³ ì„œì—ì„œ ê·¸ë˜í”„ ë§í¬ ì¶”ì¶œ
with open(report_path, 'r', encoding='utf-8') as f:
    report_content = f.read()

# Markdown ì´ë¯¸ì§€ íŒ¨í„´: ![alt](path)
image_pattern = r'!\[.*?\]\((.*?)\)'
image_links = re.findall(image_pattern, report_content)

print("ë³´ê³ ì„œ ë‚´ ê·¸ë˜í”„ ë§í¬ ê²€ì¦:")
print("="*80)

for i, link in enumerate(image_links, 1):
    # ìƒëŒ€ ê²½ë¡œë¥¼ ì ˆëŒ€ ê²½ë¡œë¡œ ë³€í™˜
    if link.startswith('../'):
        abs_path = os.path.join(DATA_PATHS['reports'], link)
        abs_path = os.path.normpath(abs_path)
    else:
        abs_path = link
    
    exists = os.path.exists(abs_path)
    status = "âœ… ì¡´ì¬" if exists else "âŒ ì—†ìŒ"
    
    print(f"{i:2d}. {link:50s} | {status}")

print(f"\nì´ {len(image_links)}ê°œ ì´ë¯¸ì§€ ë§í¬")
````

## 7. í”„ë¡œì íŠ¸ íŒŒì¼ ëª©ë¡ ìµœì¢… ì •ë¦¬

````python
import os

def list_files_recursive(directory, indent=0):
    """ë””ë ‰í† ë¦¬ êµ¬ì¡°ë¥¼ íŠ¸ë¦¬ í˜•íƒœë¡œ ì¶œë ¥"""
    items = []
    try:
        entries = sorted(os.listdir(directory))
        for entry in entries:
            # ìˆ¨ê¹€ íŒŒì¼ ë° ìºì‹œ ì œì™¸
            if entry.startswith('.') or entry == '__pycache__':
                continue
            
            path = os.path.join(directory, entry)
            prefix = "  " * indent + "â”œâ”€ "
            
            if os.path.isdir(path):
                items.append(f"{prefix}{entry}/")
                items.extend(list_files_recursive(path, indent + 1))
            else:
                file_size = os.path.getsize(path)
                size_str = f" ({file_size:,} bytes)" if file_size > 0 else ""
                items.append(f"{prefix}{entry}{size_str}")
    except PermissionError:
        pass
    
    return items

# í”„ë¡œì íŠ¸ ë£¨íŠ¸ ë””ë ‰í† ë¦¬
project_root = os.path.dirname(os.path.dirname(os.path.abspath(os.getcwd())))

print("í”„ë¡œì íŠ¸ íŒŒì¼ êµ¬ì¡°:")
print("="*80)
print("cctvë¶„ì„/")

# ì£¼ìš” ë””ë ‰í† ë¦¬ë§Œ ì¶œë ¥
for folder in ['data', 'notebooks', 'results', 'utils', 'reviews']:
    folder_path = os.path.join(project_root, folder)
    if os.path.exists(folder_path):
        print(f"â”œâ”€ {folder}/")
        for item in list_files_recursive(folder_path, 1):
            print(item)

# ë£¨íŠ¸ íŒŒì¼
root_files = ['README.md', 'SRS.md', 'PLAN.md', 'requirements.txt', '.gitignore']
for file in root_files:
    file_path = os.path.join(project_root, file)
    if os.path.exists(file_path):
        file_size = os.path.getsize(file_path)
        print(f"â”œâ”€ {file} ({file_size:,} bytes)")
````
